```python
from matplotlib import pyplot as plt
from ml_dcs.internal.ml.prediction_methods import (
    RandomForestPrediction,
    GradientBoostingPrediction,
    DecisionTreePrediction,
    LogisticRegressionPrediction,
)
from ml_dcs.domain.ml import MLCalculationTimePredictionInput2
from ml_dcs.usecase.evaluator import RandomStateEvaluator

# graph data
evaluator_rf = RandomStateEvaluator(
    input_dir_path="./tmp/2024-08-11/input",
    ml_input_class=MLCalculationTimePredictionInput2,
    prediction_class=RandomForestPrediction,
)
r2_scores_rf, _ = evaluator_rf.get_graph_elements()

evaluator_gb = RandomStateEvaluator(
    input_dir_path="./tmp/2024-08-11/input",
    ml_input_class=MLCalculationTimePredictionInput2,
    prediction_class=GradientBoostingPrediction,
)
r2_scores_gb, _ = evaluator_gb.get_graph_elements()

evaluator_dt = RandomStateEvaluator(
    input_dir_path="./tmp/2024-08-11/input",
    ml_input_class=MLCalculationTimePredictionInput2,
    prediction_class=DecisionTreePrediction,
)
r2_scores_dt, _ = evaluator_dt.get_graph_elements()

evaluator_lr = RandomStateEvaluator(
    input_dir_path="./tmp/2024-08-11/input",
    ml_input_class=MLCalculationTimePredictionInput2,
    prediction_class=LogisticRegressionPrediction,
)
r2_scores_lr, _ = evaluator_dt.get_graph_elements()

# graph settings
plt.xlabel("The value of seed")
plt.ylabel("Coefficient of determination $R^2$")
plt.xlim(0, 3000)
plt.ylim(-0.5, 1)
plt.plot([0, 3000], [0, 0], color="black", linestyle="--")

fig, (axe_1, axe_2, axe_3, axe_4) = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

axe_1.scatter(
    RandomStateEvaluator.random_states,
    r2_scores_rf,
    c="white",
    edgecolors="blue",
    s=10,
)
axe_2.scatter(
    RandomStateEvaluator.random_states,
    r2_scores_gb,
    c="white",
    edgecolors="blue",
    s=10,
)
axe_3.scatter(
    RandomStateEvaluator.random_states,
    r2_scores_dt,
    c="white",
    edgecolors="blue",
    s=10,
)
axe_4.scatter(
    RandomStateEvaluator.random_states,
    r2_scores_lr,
    c="white",
    edgecolors="blue",
    s=10,
)

plt.show()
```
