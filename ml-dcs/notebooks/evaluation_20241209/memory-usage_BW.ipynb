```python
from ml_dcs.config import config  # NOQA

INPUT_DIR = "./tmp/evaluation/20241125-scenario-BW/output"
SIMPLE_INPUT_PATH = (
    f"{INPUT_DIR}/simple/memory-usage/20241209-111211/testing-result_DT.json"
)
GNN_INPUT_PATH_L1 = "tmp/evaluation/20241125-scenario-BW/output/gnn/memory-usage/20241209-114017/testing-result.json"
GNN_INPUT_PATH_L2 = f"{INPUT_DIR}/gnn/memory-usage/20241209-115410/testing-result.json"
GNN_INPUT_PATH_L3 = "tmp/evaluation/20241125-scenario-BW/output/gnn/memory-usage/20241209-122213/testing-result.json"
X_LABEL = "Actual values (MB)"
Y_LABEL = "Predicted values (MB)"
ORDER_OF_MAG = 7
LIM = (-0.15 * 10**ORDER_OF_MAG, 4 * 10**ORDER_OF_MAG)
```

```python
import json
from ml_dcs.internal.graph.graph import GraphUtil
from ml_dcs.domain.ml_simple import MLSimpleTestingResultSet
from ml_dcs.domain.graph import Graph, PredictionAccuracyGraphData

with open(SIMPLE_INPUT_PATH, "r") as f:
    data_dict = json.load(f)
data_model = MLSimpleTestingResultSet(**data_dict)
print(data_model.model_dump_json(indent=2, by_alias=True))
data = PredictionAccuracyGraphData.from_ml_simple_class_div1000(data_model)
graph = Graph(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL,
    x_lim=LIM,
    y_lim=LIM,
    order_of_mag=ORDER_OF_MAG,
)
util = GraphUtil(graph)
util.show()
```
