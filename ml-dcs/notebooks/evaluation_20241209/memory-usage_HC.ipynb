```python
from ml_dcs.config import config  # NOQA

SIMPLE_INPUT_PATH = "tmp/evaluation/20241125-scenario-ArtGallery/output/simple/memory-usage/20241209-085526/testing-result_RF.json"
GNN_INPUT_PATH_L1 = "tmp/evaluation/20241125-scenario-ArtGallery/output/gnn/memory-usage/20241209-092253/testing-result.json"
GNN_INPUT_PATH_L2 = "tmp/evaluation/20241125-scenario-ArtGallery/output/gnn/memory-usage/20241209-092840/testing-result.json"
GNN_INPUT_PATH_L3 = "tmp/evaluation/20241125-scenario-ArtGallery/output/gnn/memory-usage/20241209-095506/testing-result.json"
X_LABEL = "Actual values (MB)"
Y_LABEL = "Predicted values (MB)"
ORDER_OF_MAG = 4
LIM = (-0.15 * 10**ORDER_OF_MAG, 2.2 * 10**ORDER_OF_MAG)
```

```python
import json
from ml_dcs.internal.graph.graph import GraphUtil
from ml_dcs.domain.ml_simple import MLSimpleTestingResultSet
from ml_dcs.domain.graph import Graph, PredictionAccuracyGraphData

with open(SIMPLE_INPUT_PATH, "r") as f:
    data_dict = json.load(f)
data_model = MLSimpleTestingResultSet(**data_dict)
print(data_model.model_dump_json(indent=2, by_alias=True))
data = PredictionAccuracyGraphData.from_ml_simple_class_div1000(data_model)
graph = Graph(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL,
    x_lim=LIM,
    y_lim=LIM,
    order_of_mag=ORDER_OF_MAG,
)
util = GraphUtil(graph)
util.show()
```
