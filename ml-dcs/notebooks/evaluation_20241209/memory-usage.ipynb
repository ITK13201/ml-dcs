```python
from ml_dcs.config import config  # NOQA

SIMPLE_INPUT_PATH = (
    lambda algorithm: f"tmp/evaluation/20241125-threshold/output/simple/memory-usage/20241209-072227/testing-result_{algorithm}.json"
)
GNN_INPUT_PATH_L1 = "tmp/evaluation/20241125-threshold/output/gnn/memory-usage/20241209-075658/testing-result.json"
GNN_INPUT_PATH_L2 = "tmp/evaluation/20241125-threshold/output/gnn/memory-usage/20241209-082106/testing-result.json"
GNN_INPUT_PATH_L3 = "tmp/evaluation/20241125-threshold/output/gnn/memory-usage/20241209-083912/testing-result.json"
X_LABEL = "Actual values (KB)"
Y_LABEL = "Predicted values (KB)"
ORDER_OF_MAG = 7
LIM = (-0.15 * 10**ORDER_OF_MAG, 1.75 * 10**ORDER_OF_MAG)
```

```python
import json
from ml_dcs.internal.graph.graph import GraphUtil
from ml_dcs.domain.ml_simple import MLSimpleTestingResultSet
from ml_dcs.domain.graph import Graph, PredictionAccuracyGraphData

with open(SIMPLE_INPUT_PATH("RF"), "r") as f:
    data_dict = json.load(f)
data_model = MLSimpleTestingResultSet(**data_dict)
print(data_model.model_dump_json(indent=2, by_alias=True))
data = PredictionAccuracyGraphData.from_ml_simple_class(data_model)
graph = Graph(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL,
    x_lim=LIM,
    y_lim=LIM,
    order_of_mag=ORDER_OF_MAG,
)
util = GraphUtil(graph)
util.show()
```
