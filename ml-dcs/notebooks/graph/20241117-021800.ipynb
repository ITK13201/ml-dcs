```python
import json
import os.path

from ml_dcs.config import config # NOQA
from ml_dcs.internal.graph.graph import GraphUtil

INPUT_DIR = "./tmp/prod-results/20241117-021800/"
X_LABEL_CT = "Actual values (ms)"
Y_LABEL_CT = "Predicted values (ms)"
X_LABEL_MU = "Actual values (KiB)"
Y_LABEL_MU = "Predicted values (KiB)"
```

```python
from ml_dcs.domain.ml_gnn import GNNTestingResult
from ml_dcs.domain.graph import Graph, PredictionAccuracyGraphData

input_path = f"{INPUT_DIR}/gnn/calculation-time/testing-result.json"

with open(input_path, "r") as f:
    data_dict = json.load(f)
data_model = GNNTestingResult(**data_dict)
data = PredictionAccuracyGraphData.from_ml_gnn_class(data_model)
mag = 6
graph = Graph(
    data=data,
    x_label=X_LABEL_CT,
    y_label=Y_LABEL_CT,
    x_lim=(-0.2 * 10**mag, 6 * 10**mag),
    y_lim=(-0.5 * 10**mag, 6 * 10**mag),
    order_of_mag=6
)
util = GraphUtil(graph)
util.show()
```
