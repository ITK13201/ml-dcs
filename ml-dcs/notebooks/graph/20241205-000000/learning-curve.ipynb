```python
from ml_dcs.config import config  # NOQA

INPUT_DIR = "./tmp/evaluation/20241125-threshold/output"
X_LABEL = "Epoch"
Y_LABEL = lambda unit: f"MSE Loss ({unit})"
Y1_LEGEND = "train"
Y2_LEGEND = "validation"
```

```python
import json
from ml_dcs.internal.graph.graph import Graph2Util
from ml_dcs.domain.ml_gnn import GNNTrainingResult
from ml_dcs.domain.graph import Graph2, LearningCurveGraphData

input_path = f"{INPUT_DIR}/gnn/calculation-time/20241206-165509/training-result.json"
mag = 11

with open(input_path, "r") as f:
    data_dict = json.load(f)
data_model = GNNTrainingResult(**data_dict)
data = LearningCurveGraphData.from_ml_gnn_training_result(data_model)
graph = Graph2(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL("ms$^2$"),
    y1_legend=Y1_LEGEND,
    y2_legend=Y2_LEGEND,
    y_lim=[-0.1 * 10**mag, 0.75 * 10**mag],
    order_of_mag=mag,
)
util = Graph2Util(graph)
util.show()
```
