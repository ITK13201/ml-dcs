```python
from ml_dcs.config import config # NOQA

INPUT_DIR = "./tmp/evaluation/20241125-threshold/output"
GNN_INPUT_PATH = f"{INPUT_DIR}/gnn/memory-usage/20241207-101044/testing-result.json"
SIMPLE_INPUT_PATH = lambda algorithm: f"{INPUT_DIR}/simple/memory-usage/20241206-155005/testing-result_{algorithm}.json"
X_LABEL = "Actual values (KB)"
Y_LABEL = "Predicted values (KB)"
ORDER_OF_MAG = 7
LIM = (-0.15 * 10**ORDER_OF_MAG, 1.75 * 10**ORDER_OF_MAG)
```

```python
import json
from ml_dcs.internal.graph.graph import GraphUtil
from ml_dcs.domain.ml_gnn import GNNTestingResult
from ml_dcs.domain.graph import Graph, PredictionAccuracyGraphData

with open(GNN_INPUT_PATH, "r") as f:
    data_dict = json.load(f)
data_model = GNNTestingResult(**data_dict)
print(data_model.model_dump_json(indent=2, by_alias=True))
data = PredictionAccuracyGraphData.from_ml_gnn_class(data_model)
graph = Graph(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL,
    x_lim=LIM,
    y_lim=LIM,
    order_of_mag=ORDER_OF_MAG
)
util = GraphUtil(graph)
util.show()
```
