```python
from ml_dcs.config import config # NOQA

INPUT_DIR = "./tmp/evaluation/20241125-threshold/output"
GNN_INPUT_PATH = f"{INPUT_DIR}/gnn/memory-usage/20241206-175104/testing-result.json"
SIMPLE_INPUT_PATH = f"{INPUT_DIR}/simple/memory-usage/20241206-155005/testing-result_GBDT.json"
X_LABEL = "Actual values (ms)"
Y_LABEL = "Predicted values (ms)"
```

```python
from ml_dcs.domain.graph import PredictionAccuracyGraphData, Graph
from ml_dcs.internal.graph.graph import GraphUtil
from ml_dcs.usecases.scenario_evaluator import GNNScenarioEvaluator

mag = 7
LIM = (-0.2 * 10**mag, 1.75 * 10**mag)
SCENARIO = "ArtGallery"

evaluator = GNNScenarioEvaluator(GNN_INPUT_PATH)
data_model = evaluator.evaluate(SCENARIO)
data = PredictionAccuracyGraphData.from_evaluation_result_class(data_model)
graph = Graph(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL,
    x_lim=LIM,
    y_lim=LIM,
    order_of_mag=mag
)
util = GraphUtil(graph)
util.show()
```
