```python
from ml_dcs.config import config # NOQA

INPUT_DIR = "./tmp/evaluation/20241125-000000/output"
X_LABEL = "Actual values (ms)"
Y_LABEL = "Predicted values (ms)"
```

```python
from ml_dcs.domain.graph import PredictionAccuracyGraphData, Graph
from ml_dcs.internal.graph.graph import GraphUtil
from ml_dcs.usecases.scenario_evaluator import ScenarioEvaluator

input_path = f"{INPUT_DIR}/gnn/calculation-time/20241203-200836/testing-result.json"
X_LIM = (-0.2 * 10**7, 4.5 * 10**7)
Y_LIM = (-0.2 * 10**7, 4.5 * 10**7)
SCENARIO = "ArtGallery"

evaluator = ScenarioEvaluator(input_path)
data_model = evaluator.evaluate(SCENARIO)
data = PredictionAccuracyGraphData.from_ml_gnn_class(data_model)
graph = Graph(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL,
    x_lim=X_LIM,
    y_lim=Y_LIM,
    order_of_mag=6
)
util = GraphUtil(graph)
util.show()
```
