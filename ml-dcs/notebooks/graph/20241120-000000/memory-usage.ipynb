```python
from ml_dcs.config import config # NOQA

INPUT_DIR = "./tmp/evaluation/20241116-000000/output/gnn/memory-usage/20241120-103255"
X_LABEL = "Actual values (KB)"
Y_LABEL = "Predicted values (KB)"
X_LIM = (-0.15 * 10**7, 1.1 * 10**7)
Y_LIM = (-0.15 * 10**7, 1.1 * 10**7)
```

```python
import json
from ml_dcs.internal.graph.graph import GraphUtil
from ml_dcs.domain.ml_gnn import GNNTestingResult
from ml_dcs.domain.graph import Graph, PredictionAccuracyGraphData

input_path = f"{INPUT_DIR}/testing-result.json"

with open(input_path, "r") as f:
    data_dict = json.load(f)
data_model = GNNTestingResult(**data_dict)
data = PredictionAccuracyGraphData.from_ml_gnn_class(data_model)
graph = Graph(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL,
    x_lim=X_LIM,
    y_lim=Y_LIM,
    order_of_mag=7
)
util = GraphUtil(graph)
util.show()
```
