```python
from ml_dcs.internal.ml.prediction_methods import RandomForestPrediction
from ml_dcs.domain.ml import MLMemoryUsagePredictionInput2
from ml_dcs.usecase.evaluator import RandomStateEvaluator

evaluator = RandomStateEvaluator(input_dir_path="./tmp/2024-08-11/input", ml_input_class=MLMemoryUsagePredictionInput2, prediction_class=RandomForestPrediction)
result = evaluator.evaluate()
print(result)
evaluator.plt_show()

ORDER_OF_MAGNITUDE = 7
evaluator.plt_show_at_best_accuracy(xlim=(-0.1*10**ORDER_OF_MAGNITUDE, 5.5*10**ORDER_OF_MAGNITUDE), ylim=(-0.1*10**ORDER_OF_MAGNITUDE, 5.5*10**ORDER_OF_MAGNITUDE), unit="KiB")
```
