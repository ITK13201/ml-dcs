```python
from ml_dcs.internal.ml.simple import GradientBoostingPrediction
from ml_dcs.domain.ml_simple import MLMemoryUsagePredictionInput2
from ml_dcs.usecases.simple_evaluator import RandomStateEvaluator

evaluator = RandomStateEvaluator(
    input_dir_path="./tmp/20241013-185457/input",
    ml_input_class=MLMemoryUsagePredictionInput2,
    prediction_class=GradientBoostingPrediction,
)
result = evaluator.evaluate()
print(result)
evaluator.plt_show()

ORDER_OF_MAGNITUDE = 7
evaluator.plt_show_at_best_accuracy(
    xlim=(-0.1 * 10**ORDER_OF_MAGNITUDE, 5.5 * 10**ORDER_OF_MAGNITUDE),
    ylim=(-0.1 * 10**ORDER_OF_MAGNITUDE, 5.5 * 10**ORDER_OF_MAGNITUDE),
    unit="KiB",
)
```
