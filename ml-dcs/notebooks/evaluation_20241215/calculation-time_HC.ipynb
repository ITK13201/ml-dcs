```python
from ml_dcs.config import config # NOQA

SIMPLE_INPUT_PATH = "tmp/prod-results/results-20241214/20241125_calculation-time_ArtGallery/simple/20241213-101820/testing-result_RF.json"
GNN_INPUT_PATH_L1 = "tmp/prod-results/results-20241214/20241125_calculation-time_ArtGallery/gnn-L1/20241213-102950/testing-result.json"
GNN_INPUT_PATH_L2 = "tmp/prod-results/results-20241214/20241125_calculation-time_ArtGallery/gnn-L2/20241213-104526/testing-result.json"
GNN_INPUT_PATH_L3 = "tmp/prod-results/results-20241214/20241125_calculation-time_ArtGallery/gnn-L3/20241213-111315/testing-result.json"
X_LABEL = "Actual values (s)"
Y_LABEL = "Predicted values (s)"
ORDER_OF_MAG = 2
LIM = (-0.5 * 10**ORDER_OF_MAG, 4 * 10**ORDER_OF_MAG)
```

```python
from ml_dcs.internal.graph.graph import GraphUtil
import json
from ml_dcs.domain.ml_simple import MLSimpleTestingResultSet
from ml_dcs.domain.graph import Graph, PredictionAccuracyGraphData

with open(SIMPLE_INPUT_PATH, "r") as f:
    data_dict = json.load(f)
data_model = MLSimpleTestingResultSet(**data_dict)
print(data_model.model_dump_json(indent=2, by_alias=True))
data = PredictionAccuracyGraphData.from_ml_simple_class_div1000(data_model)
graph = Graph(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL,
    x_lim=LIM,
    y_lim=LIM,
    order_of_mag=ORDER_OF_MAG
)
util = GraphUtil(graph)
util.show()
```
