```python
from ml_dcs.config import config  # NOQA

SIMPLE_INPUT_PATH = "tmp/prod-results/results-20241214/20241125_memory-usage_AT/simple/20241213-234250/testing-result_DT.json"
GNN_INPUT_PATH_L1 = "tmp/prod-results/results-20241214/20241125_memory-usage_AT/gnn-L1/20241213-235413/testing-result.json"
GNN_INPUT_PATH_L2 = "tmp/prod-results/results-20241214/20241125_memory-usage_AT/gnn-L2/20241214-004614/testing-result.json"
GNN_INPUT_PATH_L3 = "tmp/prod-results/results-20241214/20241125_memory-usage_AT/gnn-L3/20241214-005952/testing-result.json"
X_LABEL = "Actual values (MB)"
Y_LABEL = "Predicted values (MB)"
ORDER_OF_MAG = 3
LIM = (-0.5 * 10**ORDER_OF_MAG, 7 * 10**ORDER_OF_MAG)
```

```python
import json
from ml_dcs.internal.graph.graph import GraphUtil
from ml_dcs.domain.ml_simple import MLSimpleTestingResultSet
from ml_dcs.domain.graph import Graph, PredictionAccuracyGraphData

with open(SIMPLE_INPUT_PATH, "r") as f:
    data_dict = json.load(f)
data_model = MLSimpleTestingResultSet(**data_dict)
print(data_model.model_dump_json(indent=2, by_alias=True))
data = PredictionAccuracyGraphData.from_ml_simple_class_div1000(data_model)
graph = Graph(
    data=data,
    x_label=X_LABEL,
    y_label=Y_LABEL,
    x_lim=LIM,
    y_lim=LIM,
    order_of_mag=ORDER_OF_MAG,
)
util = GraphUtil(graph)
util.show()
```
