```python
from ml_dcs.config import config  # NOQA
from ml_dcs.internal.mtsa.data_utils import MTSADataUtil
import pandas as pd

DIR = "tmp/prepare-dataset/output/20241220_calculation-time"
```

```python
parsed_dataset = MTSADataUtil(DIR).get_parsed_dataset()

results_lts_name = []
results_calculation_time = []
results_lts_name_HC = []
results_calculation_time_HC = []
results_lts_name_AT = []
results_calculation_time_AT = []
results_lts_name_BW = []
results_calculation_time_BW = []
results_lts_name_CM = []
results_calculation_time_CM = []

for data in [
    *parsed_dataset.training_data,
    *parsed_dataset.validation_data,
    *parsed_dataset.testing_data,
]:
    results_lts_name.append(data.lts)
    results_calculation_time.append(data.duration_ms)
    if data.lts.startswith("ArtGallery"):
        results_lts_name_HC.append(data.lts)
        results_calculation_time_HC.append(data.duration_ms)
    elif data.lts.startswith("AT"):
        results_lts_name_AT.append(data.lts)
        results_calculation_time_AT.append(data.duration_ms)
    elif data.lts.startswith("BW"):
        results_lts_name_BW.append(data.lts)
        results_calculation_time_BW.append(data.duration_ms)
    elif data.lts.startswith("CM"):
        results_lts_name_CM.append(data.lts)
        results_calculation_time_CM.append(data.duration_ms)


# FOR TABLE
pd.DataFrame(
    {"lts_name": results_lts_name, "calculation_time": results_calculation_time}
).sort_values(by="calculation_time", ascending=False)
```

```python
pd.DataFrame(
    {
        "lts_name_HC": results_lts_name_HC,
        "calculation_time_HC": results_calculation_time_HC,
    }
).sort_values(by="calculation_time_HC", ascending=False)
```

```python
pd.DataFrame(
    {
        "lts_name_AT": results_lts_name_AT,
        "calculation_time_AT": results_calculation_time_AT,
    }
).sort_values(by="calculation_time_AT", ascending=False)
```

```python
pd.DataFrame(
    {
        "lts_name_BW": results_lts_name_BW,
        "calculation_time_BW": results_calculation_time_BW,
    }
).sort_values(by="calculation_time_BW", ascending=False)
```

```python
pd.DataFrame(
    {
        "lts_name_CM": results_lts_name_CM,
        "calculation_time_CM": results_calculation_time_CM,
    }
).sort_values(by="calculation_time_CM", ascending=False)
```
