```python
from ml_dcs.config import config  # NOQA
import json

SIMPLE_INPUT_PATH = (
    lambda algorithm: f"tmp/prod-results/results-20241220/20241220_calculation-time/simple/20241218-223051/testing-result_{algorithm}.json"
)
GNN_INPUT_PATH_L1 = "tmp/prod-results/results-20241220/20241220_calculation-time/gnn-L1/20241218-224116/testing-result.json"
GNN_INPUT_PATH_L2 = "tmp/prod-results/results-20241220/20241220_calculation-time/gnn-L2/20241218-230732/testing-result.json"
GNN_INPUT_PATH_L3 = "tmp/prod-results/results-20241220/20241220_calculation-time/gnn-L3/20241218-232205/testing-result.json"
X_LABEL = "Actual values (s)"
Y_LABEL = "Predicted values (s)"
ORDER_OF_MAG = 2
LIM = (-0.2 * 10**ORDER_OF_MAG, 3 * 10**ORDER_OF_MAG)
SCENARIO = "ArtGallery"
results_mae = []
results_rmse = []
results_mape = []
```

```python
from ml_dcs.domain.graph import PredictionAccuracyGraphData, Graph
from ml_dcs.internal.graph.graph import GraphUtil
from ml_dcs.usecases.scenario_evaluator import SimpleScenarioEvaluator

mag = 6

evaluator = SimpleScenarioEvaluator(SIMPLE_INPUT_PATH("RF"))
data_model = evaluator.evaluate(SCENARIO)
print(json.dumps({"mae": data_model.mae, "rmse": data_model.rmse}))
results_mae.append(data_model.mae)
results_rmse.append(data_model.rmse)
results_mape.append(data_model.mape)
data = PredictionAccuracyGraphData.from_evaluation_result_class_div1000(data_model)
graph = Graph(
    data=data, x_label=X_LABEL, y_label=Y_LABEL, x_lim=LIM, y_lim=LIM, order_of_mag=mag
)
util = GraphUtil(graph)
util.show()
```
